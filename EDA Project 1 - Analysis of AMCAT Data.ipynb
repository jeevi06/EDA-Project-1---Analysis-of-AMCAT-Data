##  Univariate Analysis

### PDFs, Histograms, Boxplots

```python

# Plot histograms for numerical columns
numerical_cols = ['Salary', '10percentage', '12percentage', 'collegeGPA', 'English', 'Logical', 'Quant', 'Domain']
data[numerical_cols].hist(figsize=(12, 10))
plt.show()

# Boxplot to check for outliers
sns.boxplot(data['Salary'])
plt.title('Boxplot of Salary')
plt.show()
```

### Categorical Vs Categorical 

```python
# Stacked bar plot for Gender and Specialization
pd.crosstab(data['Specialization'], data['Gender']).plot(kind='bar', stacked=True)
plt.title('Gender vs Specialization')
plt.show()
```

#### Observations:

There might be a trend showing specialization preference based on gender.

### Research Questions 

#### Test the Salary Claim

```python
# Filter data for relevant designations
fresh_grads = data[data['Designation'].isin(['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer'])]

# Check the average salary for fresh graduates
print(fresh_grads['Salary'].describe())
```

#### Relationship between Gender and Specialization

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(data['Specialization'], data['Gender'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("p-value:", p)
```

##### Conclusion:

The p-value is above 0.05, we can conclude that there is no statistically significant relationship between gender and specialization.
